```python
import pandas as pd

df = pd.read_csv('wdbc.data', header=None)
df.head()
```

```python
df.iloc[:,1].unique()
```

```python
column_labels = ['patient_id', 'diagnosis', 'mean radius', 'mean texture', 'mean perimeter', 'mean area',
'mean smoothness', 'mean compactness', 'mean concavity',
'mean concave points', 'mean symmetry', 'mean fractal dimension',
'radius error', 'texture error', 'perimeter error', 'area error',
'smoothness error', 'compactness error', 'concavity error',
'concave points error', 'symmetry error',
'fractal dimension error', 'worst radius', 'worst texture',
'worst perimeter', 'worst area', 'worst smoothness',
'worst compactness', 'worst concavity', 'worst concave points',
'worst symmetry', 'worst fractal dimension']

df.columns = column_labels
df
```

```python
df['diagnosis'].replace('B', 'Benign', inplace=True)
```

```python
df['diagnosis'].replace('M', 'Malignant', inplace=True)
df
```

```python
#split only the ones that matter
feature_df = df.iloc[:,2:]
label_df = df.iloc[:,1]
feature_df
```

```python
from sklearn.preprocessing import StandardScaler

x = feature_df.values
x = StandardScaler().fit_transform(x)
```

```python
import numpy as np
np.mean(x), np.std(x)
```

```python
from sklearn.decomposition import PCA

pca_breast = PCA(n_components=2)
principal_components_breast = pca_breast.fit_transform(x)
principal_breast_df = pd.DataFrame(data = principal_components_breast, columns = ['pc1', 'pc2'])
```

```python
pca_breast.explained_variance_ratio_
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.figure()
plt.figure(figsize=(10,10))
plt.xlabel('PC 1')
plt.ylabel('PC 2')
# plt.xticks()
# plt.yticks()
targets = ['Benign', 'Malignant']
colors = ['b', 'r']
for target, color in zip(targets, colors):
    indiceToKeep = label_df == target
    plt.scatter(principal_breast_df.loc[indiceToKeep, 'pc1'],
                principal_breast_df.loc[indiceToKeep, 'pc2'],
                c = color)
plt.legend(targets, prop={'size':15})
    
#tells us what data look like
#
```

```python
# percentage of var explained
pca = PCA().fit(x)
evr = pca.explained_variance_ratio_
print('\n'.join('PCA {:2d}: {:0.6f}'.format(k[0]+1, k[1]) for k in enumerate(evr)))
```

```python
plt.plot(np.cumsum(evr))
plt.xlabel('Num of Components')
plt.ylabel('Total Explained Variance')
```

# for Gisette Data (valid)

```python
gs = pd.read_csv('gisette_valid.data',delim_whitespace=True, header=None)
gs
```

```python
gs_label = pd.read_csv('gisette_valid.labels', header=None)
gs_label
```

```python
gs = pd.concat([gs_label, gs], axis =1)
```

```python
feature_gs_df = gs.iloc[:,1:]
label_gs_df = gs.iloc[:,0]
```

```python
from sklearn.preprocessing import StandardScaler
import numpy as np

x1 = feature_gs_df.values
x1 = StandardScaler().fit_transform(x1)

np.mean(x1),np.std(x1)
```

```python
feat_cols = ['feature'+str(i) for i in range(x1.shape[1])]
normal_gisette = pd.DataFrame(x1,columns=feat_cols)
normal_gisette
```

```python
# Choose the first 2 principal components
from sklearn.decomposition import PCA

pca_gisette = PCA(n_components=2)
principalComponents_gisette = pca_gisette.fit_transform(x1)

principal_gisette_df = pd.DataFrame(data = principalComponents_gisette
             , columns = ['principal component 1', 'principal component 2'])
print('Explained variation per principal component: {}'.format(pca_gisette.explained_variance_ratio_))
```

```python
principal_gisette_df
```

```python
import matplotlib.pyplot as plt

%matplotlib inline

plt.figure()
plt.figure(figsize=(10,10))
plt.xticks(fontsize=12)
plt.yticks(fontsize=14)
plt.xlabel('Principal Component 1',fontsize=20)
plt.ylabel('Principal Component 2',fontsize=20)
plt.title("Principal Component Analysis of Gisette Dataset",fontsize=20)
targets = [1, -1]
colors = ['b', 'r']
for target, color in zip(targets,colors):
    indicesToKeep = label_gs_df == target
    plt.scatter(principal_gisette_df.loc[indicesToKeep, 'principal component 1']
               , principal_gisette_df.loc[indicesToKeep, 'principal component 2'], c = color, s = 50)

plt.legend(targets,prop={'size': 15})
```

```python
pca_gs = PCA().fit(x1)

evr_gs = pca_gs.explained_variance_ratio_
print('\n'.join('PCA {:2d}: {:0.6f}'.format(k[0]+1, k[1]) for k in enumerate(evr_gs)))
```

```python
plt.plot(np.cumsum(evr_gs))
plt.xlabel('Number of Components')
plt.ylabel('Total Explained Variance')
```
